```python
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('brown')
nltk.download('stopwords')
```

```python
from textblob import TextBlob

for i in range(len(df)):
    if len(df['text'][i]) < 3:
        df['sentiment'][i] = 0
        continue
    temp,count = 0,0
    text = df['text'][i]

    blob = TextBlob(text)
    blob.tags           # [('The', 'DT'), ('titular', 'JJ'),
                    #  ('threat', 'NN'), ('of', 'IN'), ...]

    blob.noun_phrases   # WordList(['titular threat', 'blob',
                    #            'ultimate movie monster',
                    #            'amoeba-like mass', ...])

    for sentence in blob.sentences:
        count += 1
        temp += sentence.sentiment.polarity
    df['sentiment'][i] = round(temp/count,3)
    #if i % 100 == 0:
    print(i)
```

```python
## cont. 

for i in range(26582, len(df)):
    if len(df['text'][i]) < 3:
        df['sentiment'][i] = 0
        continue
    temp,count = 0,0
    text = df['text'][i]

    blob = TextBlob(text)
    blob.tags           # [('The', 'DT'), ('titular', 'JJ'),
                    #  ('threat', 'NN'), ('of', 'IN'), ...]

    blob.noun_phrases   # WordList(['titular threat', 'blob',
                    #            'ultimate movie monster',
                    #            'amoeba-like mass', ...])

    for sentence in blob.sentences:
        count += 1
        temp += sentence.sentiment.polarity
    df['sentiment'][i] = round(temp/count,3)
    #if i % 100 == 0:
    print(i)
```

```python
stopwords = nltk.corpus.stopwords.words("english")
stopwords.extend(['order', 'food', 'place', 'one',',','.',''])
```

```python
from nltk.tokenize import RegexpTokenizer

for i in range(len(df)):
    from collections import Counter
    text = df['text'][i].lower()
    tokenizer = RegexpTokenizer(r'\w+')
    
    
    text = tokenizer.tokenize(text)
    text = set(text) - set(stopwords)
    # Pass the split_it list to instance of Counter class.
    Counter = Counter(text)
  
    # most_common() produces k frequently encountered
    # input values and their respective counts.
    most_occur = list(Counter.most_common(20))
    l = [item[0] for item in most_occur]
    df['topics'][i] = l
    if i%100 ==0:
        print(i)
```
